```python
#Adding new column Total Marks
df['total marks']=df['math score']+df['reading score']+df['writing score']
```

```python
#Adding new column Percentage
df['percentage']=df['total marks']/300*100
```

Criteria of the grades are as follows:
95-100 : Grade A+
85-94 : Grade A
70-84 : Grade B
55-69 : Grade C
35-54 : Grade D
0-35 : Grade E


```python
#Defining function to calculate grade
def grade_calculate(scores):
    if scores >= 95 and scores <= 100:
        return 'Grade A+'
    elif scores >= 85 and scores < 95:
        return 'Grade A'
    elif scores >= 70 and scores < 85:
        return 'Grade B'
    elif scores >= 55 and scores < 70:
        return 'Grade C'
    elif scores >= 35 and scores < 55:
        return 'Grade D'
    elif scores >= 0 and scores < 35:
        return 'Grade E'
    
    
df['grades']=df['percentage'].apply(grade_calculate)
```
